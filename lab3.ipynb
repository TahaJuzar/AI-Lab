```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Automatically generate a set of colors (time slots) based on the number of tasks
def generate_colors(num_tasks):
    color_list = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#FF00FF', '#00FFFF', '#800080', '#FFA500', '#00FF7F']
    return color_list[:min(num_tasks, len(color_list))]

# Step 2: Create a graph representing tasks and their conflicts
def create_task_graph(task_conflicts):
    G = nx.Graph()
    for task1, task2 in task_conflicts:
        G.add_edge(task1, task2)
    return G

# Step 3: Visualize the graph coloring step by step
def visualize_graph(G, assignment, step, pos):
    plt.figure(figsize=(8, 8))
    node_colors = ['#ffffff'] * len(G.nodes)
    for node, color in assignment.items():
        node_colors[node] = color

    nx.draw(G, pos, with_labels=True, node_color=node_colors, font_weight='bold',
            node_size=1000, font_size=16, font_color='black', edge_color='black',
            linewidths=2, node_shape='o', edgecolors='black')

    plt.title(f"Scheduling Step {step}")
    plt.show()

# Step 4: Function to check if a time slot assignment is valid
def is_valid_color(G, node, color, assignment, constraints):
    # Check for neighbor conflicts
    for neighbor in G.neighbors(node):
        if neighbor in assignment and assignment[neighbor] == color:
            return False

    # Check for additional constraints
    if node in constraints:
        if color not in constraints[node]:
            return False

    return True

# Step 5: Minimum Remaining Values (MRV) Heuristic
def select_unassigned_node_MRV(G, assignment, colors):
    unassigned_nodes = [node for node in G.nodes if node not in assignment]
    mrv_node = None
    min_remaining_values = float('inf')

    for node in unassigned_nodes:
        remaining_values = sum(is_valid_color(G, node, color, assignment, {}) for color in colors)
        if remaining_values < min_remaining_values:
            min_remaining_values = remaining_values
            mrv_node = node

    return mrv_node

# Step 6: Degree Heuristic
def select_unassigned_node_Degree(G, assignment):
    unassigned_nodes = [node for node in G.nodes if node not in assignment]
    return max(unassigned_nodes, key=lambda node: len([n for n in G.neighbors(node) if n not in assignment]))

# Step 7: Backtracking with Heuristics
def backtracking_with_heuristics(G, assignment, heuristic, step, colors, pos, constraints):
    if len(assignment) == len(G.nodes):
        return assignment

    if heuristic == 'MRV':
        node = select_unassigned_node_MRV(G, assignment, colors)
    elif heuristic == 'Degree':
        node = select_unassigned_node_Degree(G, assignment)
    else:
        node = list(G.nodes)[len(assignment)]

    for color in colors:
        if is_valid_color(G, node, color, assignment, constraints):
            assignment[node] = color
            visualize_graph(G, assignment, step, pos)
            result = backtracking_with_heuristics(G, assignment, heuristic, step + 1, colors, pos, constraints)
            if result:
                return result
            del assignment[node]

    return None

# Step 8: Function to let user choose number of tasks, backtracking heuristic, conflicts, and constraints
def schedule_tasks_with_user_input():
    # Define task conflicts as pairs (task1, task2)
    task_conflicts = [
        (0, 1), (0, 2), (1, 2), (1, 3), (2, 4), (3, 4), (3, 5)
    ]

    # Define constraints: task -> allowed time slots
    constraints = {
        0: ['#FF0000', '#00FF00'],  # Task 0 can only be in red or green
        1: ['#0000FF', '#FFFF00'],  # Task 1 can only be in blue or yellow
        2: ['#FF00FF', '#00FFFF'],  # Task 2 can only be in magenta or cyan
        3: ['#800080'],              # Task 3 can only be in purple
        4: ['#FFA500', '#00FF7F'],  # Task 4 can only be in orange or spring green
        5: ['#FFFF00', '#00FF00'],  # Task 5 can only be in yellow or green
    }

    # Create a graph from the task conflicts
    G = create_task_graph(task_conflicts)

    # Generate colors (time slots)
    colors = generate_colors(len(G.nodes))

    # Choose a heuristic
    print("Choose a heuristic:")
    print("1. Minimum Remaining Values (MRV)")
    print("2. Degree Heuristic")
    choice = input("Enter 1 or 2: ")

    if choice == '1':
        heuristic = 'MRV'
    elif choice == '2':
        heuristic = 'Degree'
    else:
        print("Invalid choice. Defaulting to sequential assignment.")
        heuristic = None

    pos = nx.spring_layout(G, seed=42)

    # Perform backtracking with the chosen heuristic
    final_assignment = backtracking_with_heuristics(G, {}, heuristic, 1, colors, pos, constraints)
    print("Final Assignment (Time Slots):", final_assignment)

# Step 9: Run the scheduling with user input
schedule_tasks_with_user_input()
```
